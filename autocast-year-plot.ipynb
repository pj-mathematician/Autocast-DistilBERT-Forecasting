```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/kaggle/input/tf-training-autocast/tf_training.csv')
data.head()
```

```python
columnwise = data['pub_year'].value_counts().to_frame().reset_index().sort_values('index')
```

```python
columnwise['percent']  = 100*columnwise['pub_year']/columnwise['pub_year'].sum()
```

```python
columnwise = columnwise.reset_index(drop = True)
```

```python
columnwise.plot(x = 'index', y = 'percent', kind = 'bar').get_figure().savefig('name.jpg')
```
